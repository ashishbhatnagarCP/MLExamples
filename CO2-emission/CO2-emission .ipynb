```python
import pandas as pd
```

```python
df = pd.read_csv('CO2_emission.csv')
df.head()
```

```python
df.count()
```

```python
df = (df.dropna()).drop('Indicator Name', axis=1)
```

```python
df.count()
```

```python
df.head()
```

```python
new_df = df.set_index(['Country Name', 'country_code', 'Region']).stack().to_frame('co2')
new_df.index = new_df.index.set_names('Year', level=3)
new_df = new_df.reset_index()
new_df.head()
```

```python
import plotly.express as px

px.scatter_geo(
    new_df, 
    locations='country_code', 
    color='Region',
    color_continuous_scale=[(0.00, "green"),(1.00, "red")],
    hover_name='Country Name', 
    size='co2',
    animation_frame='Year', 
    projection='natural earth',
    labels=dict(color='Region'),
    title='Variations in the world map over the years (1990-2019.1)'
)
```

```python
px.ecdf(
    new_df,
    x='Year',
    y='co2',
    color='Region'
)
```

```python
px.box(
    new_df,
    x='Region',
    y='co2'
)
```
